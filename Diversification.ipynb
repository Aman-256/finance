```python
import matplotlib.pyplot as plt
import numpy as np

# Simulate number of assets
n_assets = np.arange(1, 101)  # from 1 to 100

# Assume average individual asset variance and covariance
avg_variance = 0.04  # 20% std dev -> 0.2^2 = 0.04
avg_covariance = 0.01

# Calculate portfolio variance based on formula:
# _p^2 = (1/n) * avg_variance + ((n-1)/n) * avg_covariance
portfolio_variance = (1 / n_assets) * avg_variance + ((n_assets - 1) / n_assets) * avg_covariance
portfolio_std_dev = np.sqrt(portfolio_variance)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(n_assets, portfolio_std_dev, label='Portfolio Std Dev (Risk)', linewidth=2)
plt.axhline(y=np.sqrt(avg_covariance), color='red', linestyle='--', label='Market Risk Floor (Systematic)')
plt.title('Effect of Diversification on Portfolio Risk')
plt.xlabel('Number of Assets in Portfolio')
plt.ylabel('Portfolio Standard Deviation (Risk)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
